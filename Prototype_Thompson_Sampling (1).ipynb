# Thompson Sampling Versi 1 untuk outputan print

```python
music_data = dataframe_lagu
user_behavior_data = datapengguna

class ThompsonSamplingMusicRecommender:
    def __init__(self, num_songs):
        self.num_songs = num_songs
        self.alpha = np.ones(num_songs)
        self.beta = np.ones(num_songs)

    def recommend_song(self):
        sampled_theta = np.random.beta(self.alpha, self.beta)
        recommended_song_index = np.argmax(sampled_theta)
        return music_data.iloc[recommended_song_index]

    def update_parameters(self, song_index, reward):
        if reward == 1:
            self.alpha[song_index] += 1
        else:
            self.beta[song_index] += 1

num_songs = len(music_data)
music_recommender = ThompsonSamplingMusicRecommender(num_songs)

num_iterations = len(datapengguna['nama_pengguna'])
print("Lagu yang di rekomendasikan")
for i in range(num_iterations):
    recommended_song = music_recommender.recommend_song()
    user_action = user_behavior_data.sample(n=1).iloc[0]
    song_index = music_data[music_data['track_id'] == user_action['track_id']].index[0]
    reward = 1 if user_action['action'] == 'like' else 0
    music_recommender.update_parameters(song_index, reward)

    print("===============================")
    print("Judul Lagu:", recommended_song['track_name'])
    print("Artis:", recommended_song['artists'])
    print("Genre:", recommended_song['track_genre'])
```

# Thompson Sampling versi 2 untuk outputan graf

```python
music_data = dataframe_lagu
user_behavior_data = datapengguna

class ThompsonSamplingMusicRecommender:
    def __init__(self, num_songs):
        self.num_songs = num_songs
        self.alpha = np.ones(num_songs)
        self.beta = np.ones(num_songs)

    def recommend_song(self):
        sampled_theta = np.random.beta(self.alpha, self.beta)
        recommended_song_index = np.argmax(sampled_theta)
        return music_data.iloc[recommended_song_index]

    def update_parameters(self, song_index, reward):
        if reward == 1:
            self.alpha[song_index] += 1
        else:
            self.beta[song_index] += 1

num_songs = len(music_data)
music_recommender = ThompsonSamplingMusicRecommender(num_songs)
num_iterations = len(datapengguna['nama_pengguna'])
recommendations_count = {}

for i in range(num_iterations):
    recommended_song = music_recommender.recommend_song()

    user_action = user_behavior_data.sample(n=1).iloc[0]
    song_name = recommended_song['track_name']

    if song_name in recommendations_count:
        recommendations_count[song_name] += 1
    else:
        recommendations_count[song_name] = 1
# Mengurutkan lagu berdasarkan frekuensi rekomendasi
sorted_recommendations = sorted(recommendations_count.items(), key=lambda x: x[1], reverse=True)

# Memilih 5 lagu dengan frekuensi rekomendasi tertinggi yang berbeda
top_5_recommendations = []
seen_frequencies = set()
for song_name, freq in sorted_recommendations:
    if len(top_5_recommendations) >= 5:
        break
    if freq not in seen_frequencies:
        top_5_recommendations.append((song_name, freq))
        seen_frequencies.add(freq)
```

```python
# Visualisasi
plt.figure(figsize=(10, 6))
plt.barh([song for song, _ in top_5_recommendations], [freq for _, freq in top_5_recommendations], color='skyblue')
plt.xlabel('Frekuensi Rekomendasi')
plt.ylabel('Judul Lagu')
plt.title('Lagu yang direkomendasikan')
plt.gca().invert_yaxis()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Dikosongkan untuk menyederhanakan contoh
# Assume music_data and user_behavior_data are defined elsewhere

music_data = dataframe_lagu
user_behavior_data = datapengguna

class ThompsonSamplingMusicRecommender:
    def __init__(self, num_songs):
        self.num_songs = num_songs
        self.alpha = np.ones(num_songs)
        self.beta = np.ones(num_songs)

    def recommend_song(self):
        sampled_theta = np.random.beta(self.alpha, self.beta)
        recommended_song_index = np.argmax(sampled_theta)
        return music_data.iloc[recommended_song_index]

    def update_parameters(self, song_index, reward):
        if reward == 1:
            self.alpha[song_index] += 1
        else:
            self.beta[song_index] += 1

num_songs = len(music_data)
music_recommender = ThompsonSamplingMusicRecommender(num_songs)

num_iterations = len(datapengguna['nama_pengguna'])

# Arrays to store recommendation outcomes
recommended_songs = []
liked_songs = []

for i in range(num_iterations):
    recommended_song = music_recommender.recommend_song()
    recommended_songs.append(recommended_song['track_name'])

    user_action = user_behavior_data.sample(n=1).iloc[0]
    song_index = music_data[music_data['track_id'] == user_action['track_id']].index[0]
    reward = 1 if user_action['action'] == 'like' else 0
    music_recommender.update_parameters(song_index, reward)

    if reward == 1:
        liked_songs.append(recommended_song['track_name'])

    print("===============================")
    print("Judul Lagu:", recommended_song['track_name'])
    print("Artis:", recommended_song['artists'])
    print("Genre:", recommended_song['track_genre'])
```

```python
from matplotlib_venn import venn2
import matplotlib.pyplot as plt
# Create sets
all_recommended_songs = set(recommended_songs)
liked_songs_set = set(liked_songs)

# Create Venn diagram
venn2([all_recommended_songs, liked_songs_set], ('Musik Direkomendasi', 'Musik Disukai'))
plt.title('Diagram Venn antara musik yang direkomendasi dan disukai')
plt.show()
```
